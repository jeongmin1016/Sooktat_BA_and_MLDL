```python
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import math
import gc
import pprint
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

# Train / Test 파일 불러오기

```python
train=pd.read_csv("train.csv")
```

```python
test=pd.read_csv('test.csv')
```

# OC

```python
train = pd.get_dummies(train, columns = ['OC'], drop_first=True) 
test = pd.get_dummies(test, columns = ['OC'], drop_first=True) # 가변수 변환 -> OC_open으로 변함 

train.rename(columns = {'OC_open' : 'OC'}, inplace = True)
test.rename(columns = {'OC_open' : 'OC'}, inplace = True)
```

오버 샘플링 필요

# sido

```python
train.sido=train.sido.astype('category')
train.sido
```

```python
test.sido=train.sido.astype('category')
test.sido
```

```python
# ~남 ~북을 제거, ex. 충남 -> 충 / 충북 -> 충 / 경, 전도 묶임
train["sido"] = train["sido"].str.replace("nam|buk", "")
test["sido"] = test["sido"].str.replace("nam|buk", "")

# 인천과 경기를 묶어줌
train["sido"] = train["sido"].str.replace("gyeonggi|incheon", "gyeon-in")
test["sido"] = test["sido"].str.replace("gyeonggi|incheon", "gyeon-in")
```

```python
# 세종 -> 충에 묶기
train["sido"] = train["sido"].str.replace("sejong", "choong")

# 광주 -> 전에 묶기
train['sido'] = train["sido"].str.replace("gwangju", "jeon")
```

#### 최종 결과

```python
train.sido.value_counts()
```

```python
test.sido.value_counts() # 여기서 경 -> 경남과 경북 묶은 형태
```

```python
fig, ax = plt.subplots(1, 2, figsize=(32, 8))
sns.countplot(x="sido", data=train, ax=ax[0]).set(title='train data')
sns.countplot(x="sido", data=test, ax=ax[1]).set(title='test data')
plt.show()
```

# instkind

```python
train.instkind=train.instkind.astype('category')
train.instkind
```

```python
train.instkind.value_counts()
```

```python
test.instkind.value_counts()
```

```python
# 덴탈 클리닉을 클리닉으로 포함
train['instkind'] = train["instkind"].str.replace("dental_clinic", "clinic")

# 한방병원(의원)을 모두 한방으로 묶기
train['instkind'] = train["instkind"].str.replace("traditional_hospital", "traditional")
test['instkind'] = test["instkind"].str.replace("traditional_hospital", "traditional")

train['instkind'] = train["instkind"].str.replace("traditional_clinic", "traditional")
test['instkind'] = test["instkind"].str.replace("traditional_clinic", "traditional")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(32, 8))
sns.countplot(x="instkind", data=train, ax=ax[0]).set(title='train data')
sns.countplot(x="instkind", data=test, ax=ax[1]).set(title='test data')
plt.show()
```

# opendate

```python
test["openDate"] = test["openDate"].fillna(0) # test 데이터에 결측치 하나 존재해서 0으로 채워줌
```

```python
train["openDate"] = pd.to_datetime(train["openDate"].astype("str"), format="%Y/%m/%d")
test["openDate"] = pd.to_datetime(test["openDate"].astype("int").astype("str"), format="%Y/%m/%d", errors="coerce")
```

```python
train["open_year"] = train["openDate"].dt.year
train["open_month"] = train["openDate"].dt.month
test["open_year"] = test["openDate"].dt.year
test["open_month"] = test["openDate"].dt.month

train.drop(columns="openDate", axis=1, inplace=True)
test.drop(columns="openDate", axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(32, 8))
sns.countplot(data=train, x="open_month", ax=ax[0]).set_title("Train - Month")
sns.countplot(data=test, x="open_month", ax=ax[1]).set_title("Test - Month")
plt.show()
```

# 결측치 확인 및 이상치 처리

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 7))
sns.heatmap(train.isnull(), ax=ax[0]).set_title("Train Missing")
sns.heatmap(test.isnull(), ax=ax[1]).set_title("Test Missing")
plt.show()
```

```python
missing_percentage = pd.DataFrame(columns=['column', 'null_percentage'])

for col in train.columns:
    p = round((train[col].isnull().sum() / train[col].shape[0]), 2)
    missing_percentage =  missing_percentage.append(pd.DataFrame([[col, p]], columns=['column', 'null_percentage']), ignore_index=True)

missing_percentage.set_index("column", inplace=True)
missing_percentage = missing_percentage.sort_values(by="null_percentage", ascending=False)
missing_percentage = missing_percentage[missing_percentage["null_percentage"] > 0]
missing_percentage
```

```python
missing_percentage = pd.DataFrame(columns=['column', 'null_percentage'])

for col in test.columns:
    p = round((test[col].isnull().sum() / test[col].shape[0]), 2)
    missing_percentage =  missing_percentage.append(pd.DataFrame([[col, p]], columns=['column', 'null_percentage']), ignore_index=True)

missing_percentage.set_index("column", inplace=True)
missing_percentage = missing_percentage.sort_values(by="null_percentage", ascending=False)
missing_percentage = missing_percentage[missing_percentage["null_percentage"] > 0]
missing_percentage
```

# 변수 간의 상관관계 고려

```python
%matplotlib inline
corr=train.drop(columns=['OC']).corr()
mask=np.triu(np.ones_like(corr, dtype=bool))[1:,:-1]
corr=corr.iloc[1:,:-1].copy()
fig, ax = plt.subplots(figsize=(48,48))   
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, center=0, annot=True, fmt='.2f', 
            cmap='coolwarm', annot_kws={'fontsize':10,'fontweight':'bold'}, cbar=False)
ax.tick_params(left=False,bottom=False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right',fontsize=12)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)
plt.title('Correlations between Variables\n', fontsize=16)
fig.show()
```

```python
train_corr = train.drop(columns=['OC']).corr()

s = train_corr.unstack()
df = pd.DataFrame(s[s < 1].abs().sort_values(ascending=False), columns=['corr'])   # 상관계수의 절댓값 크기순으로 정렬
df = df[df['corr'] >= 0.8] # 
df.style.background_gradient(cmap='viridis')  
```
